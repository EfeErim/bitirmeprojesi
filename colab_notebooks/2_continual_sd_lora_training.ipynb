# Stage 2: Continual SD-LoRA Training

Build the v6 continual trainer from `config/base.json` and optionally initialize the model engine.


```python
from pathlib import Path
import sys

def is_repo_root(path: Path) -> bool:
    return (path / "scripts").exists() and (path / "colab_notebooks").exists()

def repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for candidate in [cwd, *cwd.parents]:
        if is_repo_root(candidate):
            return candidate
    candidates = [
        Path("/content/bitirme projesi"),
        Path("/content/bitirmeprojesi"),
        Path("/content/aads_ulora"),
        Path("/content/drive/MyDrive/bitirme projesi"),
        Path("/content/drive/MyDrive/bitirmeprojesi"),
        Path("/content/drive/MyDrive/aads_ulora"),
    ]
    for candidate in candidates:
        if is_repo_root(candidate):
            return candidate
    scan_roots = [Path("/content"), Path("/content/drive/MyDrive")]
    for scan_root in scan_roots:
        if not scan_root.exists():
            continue
        for child in scan_root.iterdir():
            if child.is_dir() and is_repo_root(child):
                return child
    raise FileNotFoundError(
        "Repository root not found. Clone repo into Colab, then set notebook working directory to the cloned repo."
    )

ROOT = repo_root()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print(f"Repository root: {ROOT}")
```

```python
import json
from pprint import pprint

from src.training.continual_sd_lora import ContinualSDLoRAConfig, ContinualSDLoRATrainer

cfg_path = ROOT / "config" / "base.json"
payload = json.loads(cfg_path.read_text(encoding="utf-8"))
training_cfg = payload.get("training", {}).get("continual", {})

config = ContinualSDLoRAConfig.from_training_config(training_cfg)
print("Loaded continual training contract:")
pprint(config.as_contract_dict())

trainer = ContinualSDLoRATrainer(config)
print("Trainer created on device:", trainer.device)
```

```python
RUN_ENGINE_INIT = False
CLASS_NAMES = ["healthy_example", "disease_example"]

if RUN_ENGINE_INIT:
    class_to_idx = {name: idx for idx, name in enumerate(CLASS_NAMES)}
    trainer.initialize_engine(class_to_idx=class_to_idx)
    trainer.setup_optimizer()
    print("Engine initialized. Target modules:", len(trainer.target_modules_resolved))
else:
    print("Skipping engine initialization to avoid heavyweight model download.")
    print("Set RUN_ENGINE_INIT=True when runtime is ready.")
```

```python
SAVE_ADAPTER = False
OUTPUT_DIR = ROOT / "outputs" / "colab"

if SAVE_ADAPTER:
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
    adapter_path = trainer.save_adapter(str(OUTPUT_DIR))
    print("Adapter saved to:", adapter_path)
else:
    print("Skipping adapter save. Initialize engine first and set SAVE_ADAPTER=True.")
```
